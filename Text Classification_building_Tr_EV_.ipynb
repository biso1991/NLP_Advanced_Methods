# Sentiment Analysis on Movie Reviews
This code demonstrates the process of creating, training and evaluating sentiment analysis models for movie reviews. It uses common natural language processing techniques and provides visualizations to interpret model performance.



```python
import numpy as np
import pandas as pd
import zipfile
import sklearn
```

```python
zip_file_path = 'moviereviews.tsv.zip'
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall('moviereviews.tsv')
```

```python
df = pd.read_csv('moviereviews.tsv/moviereviews.tsv', sep='\t')
df.head()
```

```python
from IPython.display import Markdown, display
for i in range(5):
    display(Markdown('> '+df['review'][i]))
    print('--------------------------------------')
```

```python
len(df)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
blanks = []

for i,lb,rv in df.itertuples():  
    if type(rv)==str:            
        if rv.isspace():         
            blanks.append(i)     
        
print(len(blanks), 'blanks: ', blanks)
```

```python
df.drop(blanks, inplace=True)
len(df)
```

```python
df['label'].value_counts()
```

```python
from sklearn.model_selection import train_test_split
X = df['review']
y = df['label']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
y_train.value_counts() , y_test.value_counts() 
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC

text_clf_nb = Pipeline([('tfidf', TfidfVectorizer()),
                     ('clf', MultinomialNB()),])

text_clf_lsvc = Pipeline([('tfidf', TfidfVectorizer()),
                     ('clf', LinearSVC()),])

text_clf_nb.fit(X_train, y_train)
```

```python
predictions = text_clf_nb.predict(X_test)
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test,predictions))
```

```python
import matplotlib.pyplot as plt
conf_matrix = metrics.confusion_matrix(y_test, predictions)
labels = ['True Negative', 'False Positive', 'False Negative', 'True Positive']
sizes = conf_matrix.flatten()
colors = ['lightcoral', 'lightskyblue', 'lightgreen', 'gold']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal') 
plt.title('Confusion Matrix as Pie Chart')
plt.show()
```

```python
from sklearn.metrics  import classification_report 
print(metrics.classification_report(y_test,predictions))
```

```python
from collections import Counter

report = classification_report(y_test, predictions, output_dict=True)

# metrics from the classification report
precision = [report[label]['precision'] for label in report.keys() if label != 'accuracy']
recall = [report[label]['recall'] for label in report.keys() if label != 'accuracy']
f1_score = [report[label]['f1-score'] for label in report.keys() if label != 'accuracy']
labels = [label for label in report.keys() if label != 'accuracy']

# Plotting the metrics as a pie chart
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Precision Pie Chart
axs[0].pie(precision, labels=labels, autopct='%1.1f%%', startangle=90)
axs[0].set_title('Precision')

# Recall Pie Chart
axs[1].pie(recall, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1].set_title('Recall')

# F1 Score Pie Chart
axs[2].pie(f1_score, labels=labels, autopct='%1.1f%%', startangle=90)
axs[2].set_title('F1 Score')

plt.show()
```

```python
print(metrics.accuracy_score(y_test,predictions))
```

```python
accuracy = metrics.accuracy_score(y_test, predictions)

# Plotting the accuracy score as a pie chart
labels = ['Incorrect Predictions', 'Correct Predictions']
sizes = [len(predictions) - sum(predictions == y_test), sum(predictions == y_test)]
colors = ['lightcoral', 'lightgreen']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title('Accuracy Score as Pie Chart')
plt.show()
```
